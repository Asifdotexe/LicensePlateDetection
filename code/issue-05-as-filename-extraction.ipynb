```python
def get_image_filename(xml_filename):
    root = xet.parse(xml_filename).getroot()
    filename_image = root.find('filename').text
    return os.path.join(r'..\input\number-plates\images', filename_image)
```

```python
df['filepath'].apply(get_image_filename).tolist()
```
